```python
df_match['over'] = df_match['ball'].apply(lambda x:str(x).split(".")[0])
df_match['ball_no'] = df_match['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
df_match.head(10)
```

```python
df_match['balls_bowled'] = df_match['over'].astype('int')*6 + df_match['ball_no'].astype('int')
df_match['balls_left'] = 300 - df_match['balls_bowled']
df_match['balls_left'] = df_match['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
df_match.drop(columns=['over', 'ball', 'ball_no', 'striker', 'non_striker', 'bowler', 'runs_off_bat', 'extras', 'wides', 
                       'noballs', 'byes', 'legbyes', 'penalty', 'cricsheet_id'], inplace=True)
```

```python
df_match['total_runs_y'] = df_match['total_runs_y'].astype(int)
df_match['curr_score'] = df_match.groupby('match_id')['total_runs_y'].cumsum()
df_match
```

```python
df_match.drop(columns=['season', 'start_date', 'venue', 'innings', 'total_runs_y', 'wicket_type', 'other_wicket_type', 
                       'other_player_dismissed'], inplace=True)
```

```python
df_match
```

```python
df_match.isnull().sum()
```

```python
df_match['player_dismissed'] = df_match['player_dismissed'].fillna('0')
df_match['player_dismissed'] = df_match['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df_match['player_dismissed'] = df_match['player_dismissed'].astype('int')
```

```python
df_match['player_dismissed'] = df_match['player_dismissed'].astype(int)
df_match['wickets'] = df_match.groupby('match_id')['player_dismissed'].cumsum()
df_match['wickets_left'] = 10 - df_match['wickets']
```

```python
df_match
```

```python
df_match.isnull().sum()
```

```python
final_df = df_match.dropna()
final_df.isnull().sum()
```

```python
final_df.drop(columns=['player_dismissed', 'wickets'], inplace=True)
final_df
```

```python
final_df['runs_left'] = final_df['total_runs_x'] - final_df['curr_score']
final_df
```

```python
final_df['crr'] = (final_df['curr_score']*6) / final_df['balls_bowled']
final_df
```

```python
final_df['rrr'] = (final_df['runs_left']*6) / final_df['balls_left']
final_df
```

```python
final_df['winner'].unique()
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
final_df['result'] = final_df.apply(result, axis=1)
final_df
```

```python
df = final_df[final_df['runs_left'] >= 0]
df.shape
```

```python
df = df[df['balls_left'] != 0]
df.shape
```

```python
X = df.drop(columns='result')
y = df['result']
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
